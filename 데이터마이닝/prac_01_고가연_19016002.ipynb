## 문제1

파라미터 a와 b의 합이 0.9 인지 확인하는 함수를 구현하시오.

***결과***
```
True
True
True
True
```


````python
def check_seven(a, b):
    if round(a+b, 5) == 0.9: # 실수 비교시에 등호 조심!!!, 논리적으론 문제가 없음... 이건 컴퓨터적 문제임.
#     if 0.8999999 <= a+b <= 0.9: # 크거나 같다, 작거나 같다는 이런식으로 하는 거...
        return True
    else:
        # 0.8999999999999999로 계산되기도 함. (실수는 이상하게 계산되기도 함.)
        # 실수는 웬만하면 등호쓰면 안됨. (정수는 등호 써도 잘 나옴.)
        # 크거나 같다, 작거나 같다. 혹은 round로 대체할 것.
        return False
    pass
# 아래는 수정하지 마시오.
print(check_seven(0.1, 0.8))
print(check_seven(0.2, 0.7))
print(check_seven(0.3, 0.6))
print(check_seven(0.4, 0.5))
````
